```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../scripts')))
```

```python
df = pd.read_csv('/content/filtered_messages.csv')
```

```python
df.head()
```

```python
!pip install transformers
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", truncation=True)

# Apply sentiment analysis to each text entry in the DataFrame
df['sentiment'] = df['text'].apply(lambda x: sentiment_pipeline(x)[0])

# Filter and keep the relevant sentiment fields (e.g., label, score)
df['sentiment'] = df['sentiment'].apply(lambda x: {'label': x['label'], 'score': x['score']})

print(df[['text', 'sentiment']])
```

```python
df.to_csv('sentiment_analysis_results.csv', index=False)
```
