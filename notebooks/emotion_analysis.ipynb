```python
from transformers import pipeline
```

```python
emotionModel = pipeline('text-classification', model='SamLowe/roberta-base-go_emotions')
```

```python
import pandas as pd
df = pd.read_csv('/content/sentiment_analysis_results.csv')
df.head()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained("SamLowe/roberta-base-go_emotions")
emotionModel = AutoModelForSequenceClassification.from_pretrained("SamLowe/roberta-base-go_emotions")
```

```python
def get_emotion(text):
    # Tokenize the input text
    inputs = tokenizer(text, truncation=True, padding=True, max_length=512, return_tensors="pt")

    # Get the model's output
    outputs = emotionModel(**inputs)

    # Extract the label (emotion)
    emotion_label = outputs.logits.argmax(dim=1).item()  # Get the index of the highest score
    return emotion_label
```

```python
df['emotion'] = df['text'].apply(get_emotion)
```
